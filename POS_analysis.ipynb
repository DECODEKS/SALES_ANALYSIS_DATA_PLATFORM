```
from datetime import datetime
def createCurrentTimeFolder():
  currentDate=datetime.now()
  folderName= currentDate.strftime('%d-%m-%Y')
  return '/mnt/landing/POS/'+folderName
createCurrentTimeFolder()
```

```
path=createCurrentTimeFolder()
df=spark.read.parquet(path)
display(df)
```

<b>Remove all the rows where TransactionID, StoreID, ProductID,  is null

```
df.createOrReplaceTempView("pos")
df1=spark.sql("select * from pos where TransactionID is not null and storeID is not null and ProductID is not null ")
display(df1)
```

<b>Replace all the null price with default price 1000

```
df2=df1.fillna('1000', "Price")
display(df2)
```

<b>
Round of all the price to integer number

```
df3 = df2.selectExpr("cast(TransactionID as int) TransactionID","cast(StoreID as int) StoreID","cast(ProductID as int) ProductID","cast(Quantity as int) Quantity","cast(Price as int) Price","cast(TotalSale as int) TotalSale","cast(Timestamp as Date) TimeStamp")
display(df3)
```

<b>
Store the data back into ADLS folder Staged

```
def createCurrentTimeFolder_Staged():
  currentDate=datetime.now()
  folderName= currentDate.strftime('%d-%m-%Y')
  return '/mnt/staged/POS/'+folderName
createCurrentTimeFolder_Staged()
```

```
staged_path=createCurrentTimeFolder_Staged()
df3.write.save(path=staged_path, mode="Overwrite")
```
